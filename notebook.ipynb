```python
nombre = np.random.randn(100)
nombre2 = np.random.randn(100)
nombre
```

affichage des valeurs

```python
plt.figure(figsize=(4,4))
plt.plot(nombre, lw=2, color="black", marker=".", markersize=10, label="my values")
plt.plot(nombre2, lw=2, color="blue", marker="d", markersize=10, label="my values 2")
plt.grid()
plt.title('Nombre affiché')
plt.xlabel('échantillons')
plt.ylabel('valeur')
plt.legend()
```

#### Conditions

```python
# if condition:
#     action

condition = True 
if condition:
    print("ca marche")
else:
    print("ca marche pas")

condition = False
if condition:
    print("ca marche")
else:
    print("ca marche pas")

condition = False
if ~condition: # if not 
    print("ca marche")
else:
    print("ca marche pas")

condition = False
if not condition: # if not 
    print("ca marche")
else:
    print("ca marche pas")
```
